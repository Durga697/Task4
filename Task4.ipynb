```python
# ðŸ“Œ Task 4 - Machine Learning Model Implementation
# Predictive Model: Spam Detection using Naive Bayes + TF-IDF

# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 2. Load Full Dataset
# Download dataset: https://archive.ics.uci.edu/ml/machine-learning-databases/00228/smsspamcollection.zip
# Extract and place `SMSSpamCollection` file in the same folder as this notebook

df = pd.read_csv("SMSSpamCollection[1]", sep='\t', names=["label", "message"])
print("Dataset Sample:")
print(df.head())

print("\nDataset Size:", df.shape)

# 3. Encode Labels (ham=0, spam=1)
df['label_num'] = df['label'].map({'ham':0, 'spam':1})

# 4. Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    df['message'], df['label_num'], test_size=0.2, random_state=42, stratify=df['label_num']
)

# 5. Text Preprocessing - TF-IDF Vectorization
vectorizer = TfidfVectorizer(stop_words='english', max_features=5000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

# 6. Train Naive Bayes Model
model = MultinomialNB()
model.fit(X_train_tfidf, y_train)

# 7. Predictions
y_pred = model.predict(X_test_tfidf)

# 8. Evaluation
print("\nðŸ“Š Model Evaluation")
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=['Ham','Spam']))

# Confusion Matrix Visualization
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Ham','Spam'], yticklabels=['Ham','Spam'])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()

# 9. Prediction Demo with Custom Input
custom_messages = [
    "Congratulations! You have won free tickets. Claim now!",
    "Hey, what time is our meeting?",
    "URGENT! Your loan is overdue, pay immediately."
]

custom_tfidf = vectorizer.transform(custom_messages)
predictions = model.predict(custom_tfidf)

print("\nðŸ”® Prediction Demo:")
for msg, pred in zip(custom_messages, predictions):
    label = "Spam" if pred == 1 else "Ham"
    print(f"Message: {msg} --> Prediction: {label}")
```
